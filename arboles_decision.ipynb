```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('./dataset/wine.names') as f:
    column_names = [line.strip() for line in f if not line.startswith('#')]

unique_column_names = []
seen_column_names = set()
for column_name in column_names:
    if column_name in seen_column_names:
        suffix = 1
        new_column_name = f"{column_name}_{suffix}"
        while new_column_name in seen_column_names:
            suffix += 1
            new_column_name = f"{column_name}_{suffix}"
        column_name = new_column_name
    seen_column_names.add(column_name)
    unique_column_names.append(column_name)

data = pd.read_csv('./dataset/wine.data', header=None, names=unique_column_names)
```

```python
print(data.head(10))
print(data.info())
print(data.describe())
```

```python
import seaborn as sns

data.hist(bins=30, figsize=(20,15))
plt.show()

correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```
